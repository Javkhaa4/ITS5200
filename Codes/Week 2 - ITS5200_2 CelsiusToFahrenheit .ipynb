```python
import torch
import numpy as np
import torch.optim as optim

import matplotlib.pyplot as plt
```


DATA

```python
## inputs
t_c =  [0.5,  14.0, 15.0, 28.0, 11.0,  8.0,  3.0, -4.0,  6.0, 13.0, 21.0] 

## outputs
t_f = [35.7, 55.9, 58.2, 81.9, 56.3, 48.9, 33.9, 21.8, 48.4, 60.4, 68.4]
```

```python
t_c_t = torch.tensor ( t_c ) ##change data type to torch tensor. Torch tensor is support multidimensional arrays.
t_f_t = torch.tensor ( t_f )

print (t_c_t)
print (t_f_t)
```

```python
plt.plot (t_c_t.numpy(), t_f_t.numpy(), 'o')
plt.show()
```

Part 1: inference Function or Model

```python
def model(x, w, b):
    return w * x + b
```

Part 2: optimization objective

J = (y - y`)2

```python
## optimization objective

## MSE = mean squared error
## бодит утга болон таамагласан утга 2 ийн зөрүүг хянасанаар алдааг багасаж байгаа эсхээ мэднэ.

def loss_fn (y_pred, y_real):
    
    squared_diffs = (y_pred - y_real)**2 ## garsan utgaas her hol zuruutei bgaagaar nariivchlaliig tootsohiin tuld

    return squared_diffs.mean ()
```

Part 3: Training Loop

```python
## params - array -> w , b

def training_loop (n_epochs, params, optimizer, t_c_t, t_f_t):
    
    for epoch in range (0,n_epochs):
        
    
        y_pred = model (t_c_t, *params) ## first outcome is not close, but next one will more close
        
        loss = loss_fn (y_pred , t_f_t ) ## compare with real data t_f_t . we want loss value going down, min
        
        optimizer.zero_grad() ## we do those 3 steps everytime
        loss.backward ()
        optimizer.step () 
        
        print (loss)
    
```

```python

params = torch.tensor ([1.0, 0.0], requires_grad = True )

learning_rate = 0.001

optimizer = optim.SGD ([params], lr=learning_rate)
## https://pyimagesearch.com/2016/10/17/stochastic-gradient-descent-sgd-with-python/

result = training_loop (
                    n_epochs = 3000,
                    optimizer = optimizer, 
                    params = params,
                    t_c_t = t_c_t,
                    t_f_t = t_f_t
)
```
