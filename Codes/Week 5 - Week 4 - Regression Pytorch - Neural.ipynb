## Fix in case float64 error

```python
y_test.dtype ## sometimes torch gives you problem because of data had different data type 
```

```python
## fix data type

X_train = X_train.astype (np.float32)
X_test = X_test.astype (np.float32)
y_train = y_train.astype (np.float32)
y_test = y_test.astype (np.float32)
```

```python
X_train.dtype
X_test.dtype
y_train.dtype
y_test.dtype
```

```python
## convert data to torch from Numpy

X_train_tr = torch.from_numpy (X_train)
X_test_tr  = torch.from_numpy (X_test)
y_train_tr = torch.from_numpy (y_train)
y_test_tr  = torch.from_numpy (y_test)
```

## Normalization

* Normalization
* Standardization (*)

```python
## X_scaled = (X - x_means)/X_standard_dev

x_means      = X_train_tr.mean (0, keepdim = True )
x_deviations = X_train_tr.std (0, keepdim = True ) + epsilon
```

```python

x_deviations 
```

## Create the DataLoader

```python
train_ds = TensorDataset ( X_train_tr, y_train_tr )
```

```python
train_dl = DataLoader (train_ds, batch_size , shuffle=True )

## surgah ugugdluu shuffle hiij holij surgahdaa dahin ashiglana.
```

```python
train_dl
```

## Neural Network Architectures

```python
## input = 11, hidden layer = 8 , output = 1
## Linear Regression

class MLP_Net (nn.Module):
    ## nn class is neural network class
    ## init the class
    def __init__(self, x_means, x_deviations):
        super().__init__()
        
        self.x_means      = x_means       
        self.x_deviations = x_deviations
        
        self.linear1 = nn.Linear (11,128)
        self.act1    = nn.Sigmoid ()
        self.linear2 = nn.Linear (128, 1)
       ## self.dropout = nn.Dropout (0,25)
        
    ## perform inference
    def forward (self, x):
        
        x = (x - self.x_means)/ self.x_deviations ## standardarzation
        
        x = self.linear1 (x)
        x = self.act1 (x)
       ## x = self.dropout (x)
        y_pred= self.linear2 (x)
        
        ## return torch.round (y_pred)
        return y_pred
    
    
```

## Training Loop

```python

def training_loop (N_Epochs, model, loss_fn, opt):
    
    for epoch in range (N_Epochs):
        for xb, yb in train_dl:
            
            y_pred = model (xb)
            loss = loss_fn (y_pred, yb)
            
            opt.zero_grad ()
            loss.backward ()
            opt.step ()
        if epoch % 20 == 0:
            print (epoch, "loss=", loss)
            
        
            
```

```python

## model = LinRegNet ( x_means, x_deviations )
model = MLP_Net ( x_means, x_deviations )

opt = torch.optim.Adam ( model.parameters(), lr =learning_rate )
## optimazation - gradient descent

loss_fn = F.mse_loss 
## belen loss function iig duudaj bn. MSE

training_loop (N_Epochs, model, loss_fn, opt)
```
